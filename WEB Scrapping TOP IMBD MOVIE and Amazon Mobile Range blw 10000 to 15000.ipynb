```python
# Import useful libraries and classes.

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#html upload send the request
url="https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012"
page=requests.get(url)
page.text
soup=BeautifulSoup(page.text,'html.parser')
soup
```

```python
review_tex=[]

review_text_ele=soup.find_all(class_="lister-item-header")
for x in review_text_ele:
    review_tex.append(x.text)
print(review_tex)
```

```python
year_name=[]
review_text_year=soup.find_all(class_="lister-item-year text-muted unbold")
for x in review_text_year:
    year_name.append(x.text)
year_name
```

```python
run_time=[]
review_text_year_time=soup.find_all(class_="runtime")
for x in review_text_year_time:
    run_time.append(x.text)
run_time
```

```python
rating_=[]
review_text_year_time_rating=soup.find_all(class_="inline-block ratings-imdb-rating")
for x in review_text_year_time_rating:
    rating_.append(x.text)
rating_
```

```python
final_array=[]
for x,y,z,a in zip(review_tex,year_name,run_time,rating_):
    final_array.append({"review_tex":x,"year_name":y,"run_time":z,"rating_":a})
data=pd.DataFrame(final_array)
```

```python
data["review_tex"]=data["review_tex"].replace({"\n":""},regex=True)
```

```python
data["rating_"]=data["rating_"].replace({"\n":""},regex=True)
```

```python
data["year_name"]=data["year_name"].replace({"\(":""},regex=True)
```

```python
data["year_name"]=data["year_name"].replace({"\)":""},regex=True)
```
